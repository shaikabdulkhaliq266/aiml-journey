```python
import numpy as np
rng=np.random.default_rng(seed=42)

hours_studied=rng.integers(1,10,size=50)
attendance=rng.integers(50,100,size=50)
prev_score=rng.integers(40,95,size=50)
predicted_score=(hours_studied*5)+(attendance*0.3)+(prev_score*0.5)
dataset=np.column_stack([hours_studied,attendance,prev_score,predicted_score])
print(dataset)
    
```
