```python
print("Enter 2 numbers :")
a=int(input())
b=int(input())
print("Select the number of the operation you want to perform")
print("1.ADD")
print("2.SUB")
print("3.MUL")
print("4.DIV")
print("5.EXIT")
c=int(input())
if(c==1):
    print(a+b)
elif(c==2):
    print(a-b)
elif(c==3):
    print(a*b)
elif(c==4):
    print(a/b)
else:
    print("EXIT")
```
