```python
def grade_calculator(avg):
    if avg>=90:
        print("A+")
    elif avg>=80:
        print("A")
    elif avg>=70:
        print("B")
    elif avg>=60:
        print("C")
    elif avg>=50:
        print("D")
    elif avg>=35:
        print("E")
    else:
        print("F")

marks=[55,66,61,67,78,71]
avg=sum(marks)/len(marks)
print(marks)
grade_calculator(avg)

    
    
```
